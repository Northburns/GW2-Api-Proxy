```kotlin
%useLatestDescriptors
%use dataframe
%use kandy
```

```kotlin
// Creates a DataFrame by importing data from the "netflix_titles.csv" file.
val rawDf = DataFrame.read("customers-100.csv")

// Displays the raw DataFrame data
rawDf
```

```kotlin
rawDf
    .valueCounts(sort=false) { Company }
    .plot {
        bars {
            x(Company)
            y("count")
            fillColor(Company) {
                scale = categorical(range = listOf(Color.hex("#FF00FF"), Color.hex("#00FF00")))
            }
        }
    }
```
